### Executive Summary
> For any business, clustering is an important aspect of customer or market segmentation. As the optimal number of clusters are not known apriori, this is an unsupervised learning problem for which sklearn's k-Means algorithm is optimally suited. This algorithm can be used in various ways:
> + In capital markets, this algorithm can be used to segment different stocks based on sector, value/growth, low/high volatility etc.
> + For a business, this approach can be used to segment customers based on age, sex, income, zipcode, education, marital status, credit history. 
> + For fraud detection, this approach can be used to label behavior which is out of the ordinary. 

> As this algorithm was used to classify investment opportunities - a project protected by Non Disclosure Agreement - this project uses public datasets to demonstrate the approach used to implement the machine learning algorithm. This project consisted of two parts:
> + First, 2D synthetic data was created with a fixed number of centers but varying standard deviation. On each case, k Means fits were carried out. Then, fixing the number of centers, standard deviation was increased. On this, k Means was carried out again.
> + Second, a generic 12D dataset was created. On this the Elbow score v. Calinksi Harabaz score were computed to automate the identification of the optimal number of centers.

> **Conclusions**
> Even with noisy data, kMeans algorithm was successful in classifying 2D data. With large dimensional datasets (more than 3), the Calinksi Harabaz score can be used to automate the identification of the optimal number of centers.

# 1. Setting Up

### Importing libraries

```python
#!pip3 install yellowbrick
# https://www.scikit-yb.org/en/latest/api/cluster/elbow.html

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.datasets.samples_generator import make_blobs
from yellowbrick.cluster import KElbowVisualizer
```

# 2. Modeling

### 2.1. Varying cluster standard deviation

**Standard Deviation: 0.5**

```python
StdDev = 0.5
Clusters = 4

X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=StdDev, random_state=0)

kmeans = KMeans(n_clusters=Clusters)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.figure(figsize = (10,10))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Dimension 1", fontsize = 15)
plt.ylabel("Dimension 2", fontsize = 15)
#plt.savefig("CE_ML_Project_08_kMeans_01.jpeg")
plt.show()
```

# 3. Automated Optimization of Clusters

### Elbow Method v. Calinski Harabaz Score Elbow Method

**Generate synthetic dataset with 8 random clusters**

```python
X, y = make_blobs(n_samples=1000, n_features=12, centers=8, random_state=42)
model = KMeans()
```

**Model fitting**

```python
# 1. Elbow method
plt.figure(figsize = (10,10))
plt.grid(False)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Neighbors", fontsize = 15)
plt.ylabel("Distortion Score", fontsize = 15)
visualizer = KElbowVisualizer(model, k=(1,15))
visualizer.fit(X)        
plt.savefig("CE_ML_Project_08_kMeans_03_Elbow.jpeg")
visualizer.show()        
plt.show()


# 2. Calinski Harabasz
plt.figure(figsize = (10,10))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("Neighbors", fontsize = 15)
plt.ylabel("Calinski Harabasz Score", fontsize = 15)
plt.grid(False)
visualizer = KElbowVisualizer(model, k=(2,15), metric='calinski_harabasz')
visualizer.fit(X)       
plt.savefig("CE_ML_Project_08_kMeans_04_CH.jpeg")
visualizer.show()   
plt.show()
```
